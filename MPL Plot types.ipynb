```python
from matplotlib import pyplot as plt
import numpy as np

unit_topics = ['Limits', 'Derivatives', 'Integrals', 'Diff Eq', 'Applications']
grades = ["A","B","C","D","F"]
As = [6, 3, 4, 3, 5]
Bs = [8, 12, 8, 9, 10]
Cs = [13, 12, 15, 13, 14]
Ds = [2, 3, 3, 2, 1]
Fs = [1, 0, 0, 3, 0]

x = range(5)


c_bottom = np.add(As, Bs)
#create d_bottom and f_bottom here
d_bottom = np.add(c_bottom,Cs)
f_bottom = np.add(d_bottom,Ds)
#create your plot here
plt.figure(figsize=(10,8))

plt.bar(x, As)
plt.bar(x, Bs, bottom=As)
plt.bar(x, Cs, bottom=c_bottom)
plt.bar(x, Ds, bottom=d_bottom)
plt.bar(x, Fs, bottom=f_bottom)

ax = plt.subplot()
ax.set_xticks(range(len(unit_topics)))
ax.set_xticklabels(unit_topics)
plt.title("Grade distribution")
plt.ylabel("Number of Students")
plt.xlabel("Units")
plt.legend(grades)

plt.show()
#plt.savefig("my_stacked_bar.png")
plt.close()

```

```python
from matplotlib import pyplot as plt

months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]

plt.plot(months, revenue, color = "grey")
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)

#setting upper and lower bands of an error margin, this is calculated based on y_axis values
y_upper = [1.1*i for i in revenue] #defining upper bound as +10%, therefore 1.1*i (I refers to the value on y-axis), it is 
#possible to use [i+0.1*i...] as it will return same result (110%) o the upper band
y_lower = [i - 0.1 * i for i in revenue] #defining lower band as -10%, or 100% of y_value less 10%
plt.fill_between(months, y_lower, y_upper, color = "grey", alpha = 0.1)
plt.show()
plt.close()
```
