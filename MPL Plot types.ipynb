```python
from matplotlib import pyplot as plt

drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]

plt.bar(range(len(drinks)), sales1)
plt.bar(range(len(drinks)), sales2, bottom=sales1)

ax = plt.subplot() 
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks) 
plt.legend(["shop 1 sales", "shop 2 sales"])


plt.show()
plt.close()
```

```python
from matplotlib import pyplot as plt

drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]
error = [0.6, 0.9, 0.4, 0, 0.9, 0]

# Plot the bar graph here
plt.bar(range(len(drinks)), ounces_of_milk, yerr = error, capsize = 15)


ax = plt.subplot()
plt.ylabel("Ounces of milk")
plt.title("Amount of milk (oz) in each coffee drink")
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(drinks)
plt.show()
plt.close()
```

```python
from matplotlib import pyplot as plt

months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]

plt.plot(months, revenue, color = "grey")
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)

#setting upper and lower bands of an error margin, this is calculated based on y_axis values
y_upper = [1.1*i for i in revenue] #defining upper bound as +10%, therefore 1.1*i (I refers to the value on y-axis), it is 
#possible to use [i+0.1*i...] as it will return same result (110%) o the upper band
y_lower = [i - 0.1 * i for i in revenue] #defining lower band as -10%, or 100% of y_value less 10%
plt.fill_between(months, y_lower, y_upper, color = "grey", alpha = 0.1)
plt.show()
plt.close()
```

```python
from matplotlib import pyplot as plt
import numpy as np #to plot a pie chart, numpy needs to be imported

payment_method_names = ["Card Swipe", "Cash", "Apple Pay", "Other"]
payment_method_freqs = [270, 77, 32, 11]

#make your pie chart here

plt.pie(payment_method_freqs,                             #plotting a pie chart
        labels = payment_method_names, autopct="%0.1f%%") #setting labels and showing percentages
plt.legend(payment_method_names) #setting legend
plt.axis("equal") #making it look not weird

plt.show()
plt.close()
```
