```python
import pandas as pd
#1
inventory = pd.read_csv("petal_power_inventory.csv")
#2
print(inventory,"\n")
#3
staten_island = inventory[inventory.location == "Staten Island"]
print(staten_island, "\n")
#4
product_request = staten_island[["location", "product_description"]]
print(product_request, "\n")
#5
seed_request = inventory[(inventory.location == "Brooklyn") & (inventory.product_type =="seeds")]
print(seed_request, "\n")
#6
inventory["in_stock"] = inventory.quantity.apply(lambda x: False if x == 0 else True)                                     

#7 Petal Power wants to know how valuable their current inventory is. 
#Create a column called total_value that is equal to price multiplied by quantity.
inventory["total_value"] = inventory.quantity * inventory.price

#8 create a new column in inventory called full_description that has the complete description of each product.

inventory["full_description"] = inventory[["product_type", "product_description"]].apply(
    lambda row:'{} - {}'.format(row.product_type, row.product_description), axis=1)  

print(inventory, "\n")
```
