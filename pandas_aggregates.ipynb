```python
import pandas as pd
import numpy as np
orders = pd.read_csv("pandas_aggregates.csv")
print(orders.head(), "\n")

#find most expensive price
most_expensive = orders.price.max()
print(most_expensive, "\n")

#number of (unique) shoe colors in the inventory
num_colors = orders.shoe_color.nunique()
print(num_colors, "\n")

#most expensive shoes by shoe_type
pricey_shoes = orders.groupby("shoe_type").price.max()
print(pricey_shoes, "\n")

#returning a nice looking DataFrame using .reset_index() 
pricey_shoes = orders.groupby('shoe_type').price.max().reset_index()
print(pricey_shoes, "\n")

#calculate the 25th percentile for shoe price for each shoe_color
cheap_shoes= orders.groupby("shoe_color").price.apply(lambda x: np.percentile(x, 25)).reset_index()
print(cheap_shoes, "\n")

#Create a DataFrame with the total number of shoes of each shoe_type/shoe_color combination purchased. 
#Save it to the variable shoe_counts.

shoe_counts = orders.groupby(["shoe_type","shoe_color"])["id"].count().reset_index()
print(shoe_counts, "\n")

#create shoe_counts PIVOT Table
shoe_counts_pivot = shoe_counts.pivot(
    columns = "shoe_color",
    index = "shoe_type",
    values = "id"
).reset_index() #looks better without reset_index though...
print(shoe_counts_pivot, "\n")
```
