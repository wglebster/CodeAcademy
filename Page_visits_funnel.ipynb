```python
'''A CodeAcademy mini Project:
Page Visits Funnel:
Cool T-Shirts Inc. has asked you to analyze data on visits to their website. Your job is to build a funnel, which is a description of how many people continue to the next step of a multi-step process.

In this case, our funnel is going to describe the following process:

A user visits CoolTShirts.com
A user adds a t-shirt to their cart
A user clicks “checkout”
A user actually purchases a t-shirt'''

#Final thoughts after completing this project: 
#I think it would've been much easier to merge all datasets at the start and then perform calculations
#I may've miscalculated some results but not using (~ .isnull()), but I cannot re-do it right now, because if I start 
#re-calculating, I will spend anoter two hours on it, but it is quite late and I want to move on.
#I THINK I GOT THE CONCEPTS ALLRIGHT!
```

```python
import pandas as pd

#create variables and import data into variables: 
#visits lists all of the users who have visited the website
#cart lists all of the users who have added a t-shirt to their cart
#checkout lists all of the users who have started the checkout
#purchase lists all of the users who have purchased a t-shirt

visits = pd.read_csv("Page_visits_funnel_visits.csv",  parse_dates = ["visit_time"])
cart = pd.read_csv("Page_visits_funnel_cart.csv", parse_dates = [1])
checkout = pd.read_csv("Page_visits_funnel_checkout.csv", parse_dates = [1])
purchase = pd.read_csv("Page_visits_funnel_purchase.csv", parse_dates = [1])

#print(visits.head())
#print(cart.head())
#print(checkout.head())
#print(purchase.head())

#Combine visits and cart using a left merge.
visits_cart_merge = visits.merge(cart, how = "left")
print(visits_cart_merge.head(), "\n")

#How long is your merged DataFrame?
total_visits = len(visits_cart_merge["visit_time"])
print(total_visits, " was the total number of page visits \n") 

#How many of the timestamps are null for the column cart_time?
no_cart = len(visits_cart_merge[visits_cart_merge.cart_time.isnull()])
print(no_cart, " visitors left without putting anything into their shopping cart \n")

#What percent of users who visited Cool T-Shirts Inc. ended up not placing a t-shirt in their cart?
no_cart_percentage = no_cart/total_visits
print("{:.2%}".format(no_cart_percentage) + " of visitors left the page without putting anything in the cart. \n")

#Repeat the left merge for cart and checkout and count null values. 
#What percentage of users put items in their cart, but did not proceed to checkout?
cart_checkout_merge = cart.merge(checkout, how = "left")
print(cart_checkout_merge.head(), "\n")

total_cart = len(cart_checkout_merge["cart_time"])
print(total_cart, "\n")

no_checkout = len(cart_checkout_merge[cart_checkout_merge.checkout_time.isnull()])
print(no_checkout, "\n")

no_checkout_percentage = no_checkout/total_cart
print("{:.2%}".format(no_checkout_percentage) + " of those who had stuff in the cart, have abandoned without \
completing a purchase. \n")

#Merge all four steps of the funnel, in order, using a series of left merges. Save the results to the variable all_data.
all_data = visits.merge(cart, how = "left").merge(checkout, how = "left").merge(purchase, how = "left")
#print(all_data.head(), "\n")

#What percentage of users proceeded to checkout, but did not purchase a t-shirt?
total_check_out = len(all_data[~all_data.checkout_time.isnull()])
print(total_check_out, " is the number of visitors who reached the check out \n")
no_purchase = len(all_data[~all_data.purchase_time.isnull()])
print(no_purchase, " is the number of visitors who left check out without a purchase \n")
no_purchase_percentage = no_purchase/total_check_out
print("{:.2%}".format(no_purchase_percentage) + " left without a purchase \n")

#Using DataFrame all_data, calculate the average time from initial visit to final purchase.
all_data["time_to_purchase"] = all_data.purchase_time - all_data.visit_time


print(all_data.time_to_purchase, "\n")
print(all_data.time_to_purchase.mean(), "\n")
```
